```python
import matplotlib.pyplot as plt
import random
import numpy as np
```

```python
experimentos = 10
portas = [1,2,3]

fig, axes = plt.subplots(1, 3, figsize=(15,6), sharey=True)
for ax in axes:
   taxa_vitoria_troca = []
   taxa_vitoria = []
   total_experimentos = []

   for n in np.arange(1, experimentos):
      vitoria_troca = 0
      vitoria = 0    
      for _ in range(n):
         porta_escolha = random.choice(portas)
         porta_aberta = random.choice(portas)

         if porta_escolha == porta_aberta: ## Se escolhi o prêmio de primeira
            vitoria_troca +=1
         else: ## Se não escolhi o prêmio, escolhi cabra, e se escolhi cabra, a minha troca me dá o prêmio
            vitoria += 1
      
      taxa_vitoria_troca.append(vitoria_troca / n)
      taxa_vitoria.append(vitoria / n)
      total_experimentos.append(n)

   media_vitoria_troca = np.mean(taxa_vitoria_troca)
   media_vitoria = np.mean(taxa_vitoria)

   ax.plot(total_experimentos, taxa_vitoria_troca, label='Com troca')
   ax.set_title(f'{experimentos} experimentos')
   ax.axhline(y=media_vitoria_troca, color='r', linestyle='--', label=f'Taxa de vitória com troca ~{media_vitoria_troca * 100:.0f}%')
   ax.plot(total_experimentos, taxa_vitoria, label='Sem troca', color='orange')
   ax.axhline(y=media_vitoria, color='green', linestyle='--',  label=f'Taxa de vitória sem troca ~{media_vitoria * 100:.0f}%')
   ax.set_ylim(0, 1.1)
   ax.legend()
   ax.grid()

   experimentos *= 10

fig.tight_layout()
plt.show()
```
