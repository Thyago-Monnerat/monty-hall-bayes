```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
experimentos = [10, 100, 1000]
fig, axes = plt.subplots(1, 3, figsize=(15,6), sharey=True)

for i, exp in enumerate(experimentos):
   resultado = np.random.randint(1, 4, exp)  == np.random.randint(1, 4, exp)
   vitoria = np.cumsum(resultado)  ## Se escolhi o prêmio de primeira
   vitoria_troca = np.cumsum(~resultado) ## Se não escolhi o prêmio, escolhi cabra, e se escolhi cabra, a minha troca me dá o prêmio

   experimentos_arr = np.arange(1, exp+1)

   taxa_vitoria_troca = vitoria_troca / experimentos_arr
   taxa_vitoria = vitoria / experimentos_arr
   
   media_vitoria_troca = taxa_vitoria_troca[-1]
   media_vitoria = taxa_vitoria[-1]

   axes[i].set_title(f'{exp} experimentos')
   axes[i].plot(experimentos_arr , taxa_vitoria_troca, label='Com troca')
   axes[i].axhline(y=media_vitoria_troca, color='r', linestyle='--', label=f'Taxa de vitória com troca ~{media_vitoria_troca * 100:.0f}%')
   axes[i].plot(experimentos_arr , taxa_vitoria, label='Sem troca', color='orange')
   axes[i].axhline(y=media_vitoria, color='green', linestyle='--',  label=f'Taxa de vitória sem troca ~{media_vitoria * 100:.0f}%')
   axes[i].set_ylim(0, 1.1)
   axes[i].legend()
   axes[i].grid()

fig.tight_layout()
plt.show()
```
